# Exercise 2

Create a function that will take a list of web pages and return a list of all the anchor tags on those pages.

```python
import requests

def get_links(urls):
    links = []
    for url in urls:
        res = requests.get(url)
        soup = BeautifulSoup(res.text)
        tags = soup.find_all('a')
        links += [tag for tag in tags]
    return links

list_of_urls = ["https://www.theknowledgehouse.org", 'https://www.politico.com/2020-election/results/', 'http://www.cnn.com']

get_links(list_of_urls)
```

# Exercise 3:

Create a function that accesses a web page and returns data from that page.

```python
def collect_data(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text)
    span = soup.span
    
    return span.contents[0]

    
collect_data("https://www.nba.com/")
```
