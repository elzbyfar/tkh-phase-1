# Exercise 1

Create a function that takes a url, creates a `GET` request, executes it and then returns the content of the webpage.

```python
import urllib3

def fetch_site(site_url):
    
    http = urllib3.PoolManager()
    response = http.request('GET', site_url)

    return response.data

webpage_data = fetch_site('https://google.com')
webpage_data
```

# Exercise 2

Create a function that takes two arguments (the content of a webpage and a html tag) and returns a list of all those tags from the webpage.

```python
from bs4 import BeautifulSoup as bs
```

```python
def tag_finder(webpage, tag):
    soup = bs(webpage)
    return soup.select(tag)

tag_finder(webpage_data, 'title')
```

# Exercise 3

Create a class called `Bot` that stores attributes `url`, and the methods from the previous exercises. Alter the methods to store attributes such as `status_code`, `headers` and `content` when the request is executed. Show that we're able to use this class to create instances of a bot that will scrape a website for data.

```python
from bs4 import BeautifulSoup as bs
import requests

class Bot:
    def __init__(self, url):
        self.url = url
        self.status_code = None
        self.headers = None
        self.content = None
        
    def fetch_site(self):
        response = requests.get(self.url)
        self.status_code = response.status_code
        self.headers = response.headers
        self.content = response.text
        
    def tag_finder(self, tag):
        if self.content:
            soup = bs(self.content)
            return soup.select(tag)
        else:
            self.fetch_site()
            return self.tag_finder(tag)
        
sir = Bot('https://httpbin.org/')

print("status code:", sir.status_code)
print("headers:", sir.headers)
print("content:", sir.content, '\n')

# If content is not set, tag_finder will call the fetch_site() method.
sir.tag_finder('span')

print("STATUS CODE:", sir.status_code, '\n')
print("HEADERS:", sir.headers, '\n')
print("CONTENT:", sir.content, '\n')
```
