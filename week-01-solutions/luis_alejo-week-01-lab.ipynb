## Question 2:

In your own words, explain what a float is and why sometimes it gives an incorrect value.

A float is a decimal number. Computers are very good at making fast and large calculations and can be extremely precise. However, certain floats can be the cause of errors. The reason for this, there are literally infinite possible real numbers between integers 0 and 1. An example of this is 1/3, this will return 0.333333333333---> etc.. Floating point numbers need to eventually round up in an effort to prevent an infinite calculation. As a result the precision is slightly compromised.
